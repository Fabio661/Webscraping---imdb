```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
cabecalhos = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36 OPR/94.0.0.0 (Edition std-1)'}

requests.get('https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm', headers = cabecalhos)

resposta = requests.get('https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm')
desorganizado = resposta.text
resposta
```

```python
organizado = BeautifulSoup(desorganizado, "html.parser")
```

```python
filmes = organizado.find_all("td", {"class": "titleColumn"})
filmes_avaliacoes = organizado.find_all("td", {"class": "ratingColumn imdbRating"})

filmes_nome = []
filmes_data = []
filmes_nota = []

for filme in filmes:
  nome = filme.find("a").text
  data = filme.find("span").text.strip("()") 
  filmes_nome.append(nome)
  filmes_data.append(data)

 
for filme in filmes_avaliacoes:
  try:
    nota = filme.find("strong").text
    filmes_nota.append(nota)
  except AttributeError:
    filmes_nota.append("Sem nota")
```

```python
imdb = {"Filmes": filmes_nome,
       "Data de Lancamento": filmes_data,
       "Notas": filmes_nota}

filmes_populares = pd.DataFrame(imdb)
display(filmes_populares)
```
